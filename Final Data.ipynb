```python
def spliting():
    t = ALL_COUNTRIES['publish_time']
    time = []
    for i in range(len(t)):
        ptd = t[i][:10]
        time.append(ptd)
    return(time)
```

```python
ALL_COUNTRIES['Publish_Date'] = spliting()
```

```python
def td():
    tdd = ALL_COUNTRIES['trending_date']
    date_td = []
    for i in range(len(tdd)):
        Td = tdd[i].replace('.',"")
        Tdd = ('20' + Td)
        Tds = Tdd[0:4] +'-'+ Tdd[6:]+'-'+ Tdd[4:6]
        date_td.append(Tds)
    return(date_td)
    
```

```python
ALL_COUNTRIES['Trending_Dates'] = td()
```

```python
def trend():
    D1 = ALL_COUNTRIES['Publish_Date']
    D2 = ALL_COUNTRIES['Trending_Dates']
    days= []
    for i in range(len(D1)):
        diff = np.busday_count(D1[i],D2[i], weekmask = '1111111')
        days.append(diff)
    return(days)
```

```python
ALL_COUNTRIES['Days_to_Trend'] = trend()
```

```python
m = ALL_COUNTRIES['Publish_Date']
mont= []
for i in range(len(m)):
    mh =int(m[i][5:7])
    mont.append(mh)
```

```python
def month():
    month_name = []
    for i in range(len(mont)):
        mh1 = calendar.month_name[mont[i]]
        month_name.append(mh1)
    return(month_name)
```

```python
ALL_COUNTRIES['Publish_month'] = month()
```

```python
def week_day():
    wd = ALL_COUNTRIES['Publish_Date']
    week_td = []
    for i in range(len(wd)):
        wd1 = wd[i].replace('-',"")
        #print(wd1)
        w_day = datetime.datetime(int(wd1[0:4]),int(wd1[4:6]),int(wd1[6:]))
        day_n = calendar.day_name[w_day.weekday()]
        week_td.append(day_n)
    return(week_td)
```

```python
ALL_COUNTRIES['Publish_day'] = week_day()
```
